```python
# Load the JSON file into a Python object
with open('employees_data.json', 'r') as json_file:
    loaded_data = json.load(json_file)

# Display the loaded data
print(loaded_data)
```

```python
print(loaded_data[-1])
```

```python
# Check the type of the loaded data

all_dicts = all(isinstance(item, dict) for item in loaded_data)

# Print the result
if all_dicts:
    print("All elements are dictionaries.")
else:
    print("Not all elements are dictionaries.") 
```

```python
# Add a new employee record to the loaded data
new_employee = {
    'employee_id': 104,
    'name': 'Michael Brown',
    'age': 42,
    'department': 'Sales',
    'position': 'Sales Executive',
    'salary': 90000,
    'hire_date': '2020-01-05',
    'full_time': True
}

loaded_data.append(new_employee)

# Write the modified data back to the new JSON file
with open('new_employees_data.json', 'w') as json_file:
    json.dump(loaded_data, json_file, indent=4)

# Display the updated data
print(loaded_data)
```
